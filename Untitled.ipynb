```python
import pandas as pd
import torch
import numpy as np
import string

import RNN as models
```

```python
all_letters = string.ascii_letters + " .,;'"
n_letters = len(all_letters)

def load_model(model, filename="gender_weights.pth"):
    state_dict = torch.load(filename)['state_dict']
    model.load_state_dict(state_dict)

def letterToIndex(letter):
    return all_letters.find(letter)

def lineToTensor(line):
    tensor = torch.zeros(len(line), 1, n_letters)
    for li, letter in enumerate(line):
        tensor[li][0][letterToIndex(letter)] = 1
    return tensor[:, 0]

def get_gender(name, model=None):
#     print(name)
    if(model == None):
        model = models.RNN(input_size=n_letters, hidden_size=20, num_layers=1)
        load_model(model)
    out = model(lineToTensor(name.split(' ')[0]).unsqueeze(0))
    score = out[0][0][0].data
    if score > 0.5:
        return 'Feminino'
    else:
        return 'Masculino'
```

```python
rnn = models.RNN(input_size=n_letters, hidden_size=20, num_layers=1)
load_model(rnn)
```

```python
uepb_data = pd.read_csv('07-2018 data.csv')
uepb_data['genero'] = uepb_data['nome'].apply(get_gender)

uepb_data
```
